```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

np.random.seed(1)
neu = 0.0005
std = 0.01
returns = np.random.normal(loc=neu, scale=std, size=10000)
conf_level = 0.95

#Historical
portfolio = 100
pnl = portfolio*returns
sorted_pnl = np.sort(pnl)

percentile_index = int(( 1 - conf_level )*(len(sorted_pnl)))

Var = -sorted_pnl[percentile_index]
Cvar = -np.mean(sorted_pnl[:percentile_index])

print(Var)
print(Cvar)


#Parametric
z_score = norm.ppf(1-conf_level)
Var2 = -(neu + (z_score * std ))

Cvar2 = - ( neu - ( ( norm.pdf(z_score, loc=0, scale=1) / ( 1 - conf_level )) * std ) )

print(Var2)
print(Cvar2)

plt.hist(sorted_pnl, bins=20, density=True, alpha=0.3)
plt.xlabel("Portfolio Returns")
plt.ylabel("Density")
plt.title(f"Portfolio Percentage Returns")
plt.show()
```
